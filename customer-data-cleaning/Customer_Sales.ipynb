```python
import pandas as pd
```

```python
data = pd.read_csv("/Users/ivyadiele/Downloads/customer_transactions_dirty.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df.fillna({"Payment Method": "Unknown"}, inplace=True)
df.fillna({"Country": "Unknown"}, inplace=True)
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"], errors = 'coerce')
```

```python
df.drop_duplicates(inplace = True)
```

```python
Q1 = df["Purchase Amount"].quantile(0.25)
Q3 = df["Purchase Amount"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df["Purchase Amount"] = df["Purchase Amount"].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
```
