```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data =pd.read_csv("/Users/ivyadiele/Downloads/Real_Estate_Sales_2001-2020_GL.csv")
df = pd.DataFrame(data)

df.head(10)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
Avg_AV = df.groupby('List Year')['Assessed Value'].mean().round(2)
print(Avg_AV)
```

```python
Avg_Sales = df.groupby('List Year')['Sale Amount'].mean().round(2)
print(Avg_Sales)
```

```python
Avg_ratio = df.groupby('List Year')['Sales Ratio'].mean().round(2)
print(Avg_ratio)
```

```python
df['Date Recorded'] = pd.to_datetime(df['Date Recorded'])
```

```python
df['Year Sold'] = df['Date Recorded'].dt.year
df['Month Sold'] = df['Date Recorded'].dt.month
df.head()
```

```python
Avg_yearsold = df.groupby('Property Type')['Years until sold'].mean().round(2)
print(Avg_yearsold)
```

```python
Avg_yearsoldR = df.groupby('Residential Type')['Years until sold'].mean().round(2)
print(Avg_yearsoldR)
```

```python
sales_2021 = df[df['Year Sold'] == 2021]
```

```python
town_sales_count = sales_2021.groupby('Town').size()
town_sales_count = town_sales_count.sort_values(ascending=False)
print(town_sales_count)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Property Type",y="Years until sold",data =df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x="Residential Type",y="Years until sold",data =df)
plt.show()
```
