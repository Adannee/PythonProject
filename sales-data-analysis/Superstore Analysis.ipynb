```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv("/Users/ivyadiele/Downloads/Superstore.csv", encoding="ISO-8859-1")
df = pd.DataFrame(data)
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ["Row ID", "Ship Mode", "Postal Code"], inplace = True)
```

```python
df.info()
```

```python
Total_sales = df["Sales"].sum()
print("Total Sales: $", Total_sales)
Total_profit = df["Profit"].sum()
print("Total Profit: $", Total_profit)
Total_order = df["Order ID"].nunique()
print("Total Order:", Total_order)
```

```python
top_categories = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(top_categories)
```

```python
plt.figure(figsize =(8,4))
sns.barplot(x= top_categories.index, y = top_categories.values)
plt.title("Top Performing Categories")
plt.show()
```

```python
top_products_order = df.groupby("Product Name")["Order ID"].nunique().sort_values(ascending=False).head(5)
print(top_products_order)
```

```python
plt.figure(figsize =(12,6))
sns.barplot(x= top_products_order.index, y = top_products_order.values, palette ="Blues_r")
plt.title("Top 5 most ordered Products")
plt.xlabel("Product Names")
plt.ylabel("Number of Orders")
plt.show()
```

```python
top_products_sales = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(5)
print(top_products_sales)
```

```python
top_product_profit = df.groupby("Product Name")["Profit"].sum().sort_values(ascending=False).head(5)
print("Top profitable products are: \n", top_product_profit)
```

```python
worst_product_profit = df.groupby("Product Name")["Profit"].sum().sort_values().head(5)  
print("Least Profitable Products: \n", worst_product_profit)
```

```python
top_region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False).head(5)
print("Top profitable regions are: \n", top_region_profit)
```

```python
top_region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False).head(5)
print("Top regional sales are: \n", top_region_sales)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"],  errors="coerce")  
df.set_index("Order Date", inplace=True)  
```

```python
monthly_sales = df.resample("ME").sum()["Sales"]

plt.figure(figsize=(10,5))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Months")
plt.ylabel("Sales ($)")
plt.show()
```

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending = False).head(10)
print(top_customers)
```
